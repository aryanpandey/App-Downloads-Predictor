```python
#Dropping all the columns
cols_to_drop = ['adSupported', 'androidVersionText', 'appId', 'comments', 'descriptionHTML',
                'developerEmail', 'developerId', 'developerInternalID', 'developerWebsite',
                'genreId', 'headerImage', 'icon', 'privacyPolicy', 'recentChangesHTML', 'sale',
                'saleText', 'saleTime', 'screenshots', 'summaryHTML', 'url', 'version',
                'video', 'videoImage']

data.drop(cols_to_drop, axis = 1, inplace = True)
```

```python
data.head()
```

```python
#Dealing with androidVersion
def clean_version(entry):
    if entry == entry:
        entry = str(entry)
        if entry == 'Varies':
            return -1
        elif (entry.endswith("2017") == True) or (entry.endswith("2016") == True):
            return 7
        elif entry.split('.')[1] == '4W':
            return 4
        else:
            return int(entry.split('.')[0])
    
data['androidVersion'] = data['androidVersion'].apply(clean_version)
```

```python
data.head()
```

```python
#Clean contentRating
def clean_content(entry):
    if entry == entry:
        entry = str(entry)
        if entry.startswith("Rated") == False:
            return np.nan
        else:
            return entry
        
data['contentRating'] = data['contentRating'].apply(clean_content)
```

```python
data.head()
```

```python
#Dealing with contentRatingDescription
def clean_contentRatingDescription(entry):
    if entry == entry:   
        if isinstance(entry, list) == True:
            return np.nan
        else:
            return entry
        
data['contentRatingDescription'] = data['contentRatingDescription'].apply(clean_contentRatingDescription)
```

```python
import re
```

```python
def clean_description(desc):
    if desc == desc:
        desc = re.sub(r'[^ \w\.]', '', desc).lower()
        desc = desc.replace('•', '.')
        return desc

data['description'] = data['description'].apply(clean_description)
```

```python
data['description'][0]
```

```python
def clean_free(entry):
    if entry == entry:
        if entry == "True":
            return 1
        elif entry == "False":
            return 0
        else:
            return int(float(entry))
        
data['free'] = data['free'].apply(clean_free)
```

```python
data['recentChanges'] = data['recentChanges'].apply(clean_description)
```
