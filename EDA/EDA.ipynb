From all these plots, the major inference which I am going to make here is that the data is not normalised.

For tree based models this will not make much of a difference but the other models which we try out might be heavily dependent on this normalisation. So we will keep in mind to apply a standard scalar to the data before letting our models train on it.

One more thing which I notice is that all the features have very uneven distributions. We can check to see if this uneven distribution leads to a better model or does it skew the model in some way.
